# Operators

```python
code = """
struct IntWrapper
{
    IntWrapper(int v): value(v) {}
    int value;
    
    // Spaceship operator
    int operator<=>(IntWrapper other) { return value - other.value; }
        
    IntWrapper operator+(IntWrapper other) { return IntWrapper(value + other.value); }
    IntWrapper operator+=(IntWrapper other) { value += other.value; return *this; }

    // unary minus operator
    IntWrapper operator-() { return IntWrapper(-value); }
};
"""
litgen_notebook_display.generate_and_display(options, code)
```
